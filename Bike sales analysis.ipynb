# 1. Bike Quantity by Store Location
<p>Start  by grouping the datafram so as to establish the relationship between the various categorical and numeriacls variables</p>

```python
# finding the relationship between Price and bike model
# getting the averageprice of Bikes and the sum of the Quantities
df1=df.groupby(["Store_Location"]).agg({'Price':'mean', 'Quantity':'sum'}).reset_index()
plt.figure(figsize=(7,7))
plt.title("Bike Quantity by Store Location")
sns.lineplot(x="Store_Location", y="Quantity", data=df1, orient='x')
print("Philadephia has sold the most Bikes, while Los Angeles sold the least")
```

# 2.Bike Model vs Price

exploring the relationship between the Average price of Bikes and the different Bike Models

```python
df2=df.groupby(['Bike_Model']).agg({'Price':'mean'}).reset_index().sort_values(by='Price')
plt.figure(figsize=(8,5))
sns.lineplot(x="Bike_Model", y="Price", data=df2)
plt.title("Line plot showing the relationship between Average Price and Bike Model")
print("BMX Bike_Model has the highest average price while the Folding Bike has the lowest average price")
```

# 3.BIKE MODEL vs CUSTOMER AGE: age categorized into 3 subgroups

Age will be categorized into 3 subgroups for efficiency and ofcourse to reduce the ambiguity of data. the 3 subgroups of age are:<p><ul><li>Young-35 yrs and below</ul></li>
<ul><li>Mature- between 35years and 55 years old</li></ul>
<ul><li>Old- over 55 years old</li></ul></p>

```python
# The relationship between Customer age and sum of Quantity sold
df3= df.groupby(["Customer_Age"]).agg({'Quantity':'sum'}).reset_index().sort_values(by="Customer_Age", ascending=True)
```

```python
# categorize Age to get a reduced data
def categorize_age(x):
    if x<= 35:
        return "Young"
    elif x>35 and x<= 55:
        return "Mature"
    else:
        return "Old"


# appplying the changes in the 'Customer_Age' column
df["Customer_Age"]=df["Customer_Age"].apply(lambda x:categorize_age(x))
```

```python
# Establish the relationship using the grouped age data
df4= df.groupby(["Customer_Age"]).agg({'Quantity':'sum'}).reset_index()
plt.title("Total Quantity of Bikes vs Customer Age")
sns.lineplot(x="Customer_Age", y="Quantity", data=df3)
print("The Mature population have acquired more bikes than the Young and the old")
```

```python
# Customer Age group and the Bike Model preferred
df_old=df[df.Customer_Age=='Old']['Bike_Model'].value_counts()
print("The Old Age group prefer:", df_old)
df_Young= df[df.Customer_Age=='Young']['Bike_Model'].value_counts()
print("")
print("The Young Age group prefer",df_Young)
df_mat=df[df.Customer_Age== 'Mature']['Bike_Model'].value_counts()
print("")
print("The Mature Age group prefer", df_mat)
```

```python
# plotting the prefrence bike model for the Young Age group
df5=df[df.Customer_Age=='Young']["Bike_Model"].value_counts().reset_index()
plt.figure(figsize=(8,7))
plt.title("Bike Model Prefence for Young Adults")
sns.lineplot(x= "Bike_Model", y='count', data=df5)
```

```python
# plotiing the bike preference model for the Mature age group
df7=df[df.Customer_Age=="Mature"]["Bike_Model"].value_counts().reset_index()
plt.figure(figsize=(8,6.5))
plt.title("Bike Preference Model for the Mature Age group")
sns.lineplot(x="Bike_Model", y="count",data=df7)
```

```python
# plotting the bike model preference for the Old age group
df8=df[df.Customer_Age=="Old"]["Bike_Model"].value_counts().reset_index()
plt.title("Bike Model preference plot for the Old Age group")
sns.lineplot(x="Bike_Model", y="count", data=df8)
print("The old prefere BMX Bike to Folding Bike")
```

# 4. Bike Quantity by Gender
<p>This will be to establish which Gender purchased how many bikes</p>

```python
dfgen=df.groupby(["Customer_Gender"]).agg({'Quantity':'sum'}).reset_index()
plt.title("Quantity of Bikes by Gender")
sns.lineplot(x="Customer_Gender", y="Quantity", data=dfgen)
print("The female have more bikes than the male")
```
